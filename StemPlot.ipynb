STEM PLOT

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6]
y=[2,-2,5,7,-3,6]

plt.stem(x,y, linefmt="-.", markerfmt="r*")
#solid -> -
#dashed -> --
#dash-dot -> -.
#dotted -> :
```

```python
x=[1,2,3,4,5,6]
y=[2,-2,5,7,-3,6]
plt.xlabel("Popularity")
plt.ylabel("Number")
plt.stem(x,y, linefmt="-.", markerfmt="r*", bottom = 1, basefmt = "pink", label="Python")
plt.legend()
```

```python
x=[1,2,3,4,5,6]
y=[2,-2,5,7,-3,6]
plt.xlabel("Popularity")
plt.ylabel("Number")
plt.stem(x, y, orientation="horizontal") 
plt.legend()
```
