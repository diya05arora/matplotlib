PIE PLOT

```python
import matplotlib.pyplot as plt
```

```python
x=[10,20,30,40]
y=["C", "C++", "Java", "Python"]
ex=[0.4, 0.0, 0.0, 0.0]
plt.pie(x, labels=y, explode=ex)
```

```python
x=[10,20,30,40]
y=["C", "C++", "Java", "Python"]
c=["r", "b", "g", "orange"]
plt.title("Pie Plot",fontsize=20, loc="right")
# plt.pie(x, labels=y, colors=c, autopct="%0.0f%%")
# plt.pie(x, labels=y, colors=c, autopct="%0.1f%%", shadow=True, radius=1.5, labeldistance=1.1)
# plt.pie(x, labels=y, colors=c, autopct="%0.1f%%", shadow=True, radius=1.5, startangle=90, textprops={"fontsize":15, "color":"magenta"})
# plt.pie(x, labels=y, colors=c, autopct="%0.1f%%", shadow=True, radius=1.5, counterclock=False)
# plt.pie(x, labels=y, colors=c, autopct="%0.1f%%", shadow=True, explode=ex, radius=1, wedgeprops={"linewidth":6, "width":1, "edgecolor":"pink"})
# plt.pie(x, labels=y, colors=c, autopct="%0.1f%%", shadow=True, radius=1.5, counterclock=False, rotatelabels=True)
plt.pie(x, labels=y, colors=c, autopct="%0.1f%%", shadow=True, explode=ex, radius=1.5, wedgeprops={"width":1})
plt.legend(loc=3) #or "lower left"
plt.show()
```

```python
x=[10,20,30,40]
x1=[40,30,20,10]
y=["C", "C++", "Java", "Python"]
plt.pie(x, labels=y, radius=1, colors=["red", 'pink', "yellow", "blue"])
plt.pie(x1, labels=y, radius=0.7, colors=["purple", 'green', "black", "brown"], labeldistance = 0.5, textprops={"color":"white"})
```
