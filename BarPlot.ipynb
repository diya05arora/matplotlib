BAR PLOT

```python
import matplotlib.pyplot as plt
```

```python
x = ["Python", "C", "C++", "Java"]
y = [85, 70, 60, 82]
```

```python
plt.xlabel("Languages", fontsize=15)
plt.ylabel("Popularity", fontsize=15)
plt.title("Language Users", fontsize=20)
c=["orange", "blue", "green", "red"]
#plt.bar(x,y, width=0.3, color="pink")
#plt.bar(x,y, width=0.3, color=c, align="edge", edgecolor="black", linewidth=5, linestyle="dotted", alpha=0.7) #default align in center

plt.bar(x, y, width=0.3, color=c, label="Popularity of Languages")
plt.legend() #to view the label

#ANOTHER METHOD TO USE LEGEND
# plt.bar(x,y, width=0.3, color="pink")
# plt.legend(["Popularit"])
plt.show()
```

```python
#MULTIPLE BAR GRAPHS
import numpy as np

width=0.3
z=[20,20,40,50]
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("Languages", fontsize=15)
plt.ylabel("Popularity", fontsize=15)
plt.title("Language Users", fontsize=20)

plt.bar(p,y, width, color="pink", label="Popularity of languages", edgecolor="black")
plt.bar(p1,z, width, color="red", label="Speed of languages", edgecolor="black")

#plt.xticks(position, values to be shown, rotation)
plt.xticks(p+width/2, x, rotation=20)

plt.legend(loc="upper right")
plt.show()
```

```python
#HORIZONTAL BAR GRAPH
import numpy as np

width=0.3
z=[20,20,40,50]
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("Languages", fontsize=15)
plt.ylabel("Popularity", fontsize=15)
plt.title("Language Users", fontsize=20)

plt.barh(p,y, width, color="pink", label="Popularity of languages", edgecolor="black")
plt.barh(p1,z, width, color="red", label="Speed of languages", edgecolor="black")
plt.xlabel("Popularity", fontsize=15)
plt.ylabel("Languages", fontsize=15)

#plt.yticks(position, values to be shown, rotation)
plt.yticks(p+width/2, x, rotation=20)

plt.legend(loc="upper right")
plt.show()
```
